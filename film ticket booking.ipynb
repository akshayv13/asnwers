```python
class booking:
    
    
    def __init__(self):
        
        self.matrix=[]
        self.r=int(input("Enter no of rows:"))
        self.c=int(input("Entet no of coloumns"))
        self.total_seats=self.r*self.c
        self.total_income=0
        self.userdict=dict()
        
        for i in range(self.r):
            self.l=[]
            for j in range (self.c):
                self.l.append("S")
                
            self.matrix.append(self.l)
        self.menu()
        
    def menu(self):  
        self.choice=int(input("Enter the number according to your choice\n1) Show the seats\n2) Buy a ticket\n3) Statistics\n4) booked user info\n0) Exit\n"))
         
        if self.choice ==1:
            self.seat_available() 
        elif self.choice==2:
            self.buy()
        elif self.choice==3:
            self.statistics()
        elif self.choice==4:
            self.booking_details()
        else:
            return
                    
        
    def seat_available(self):
        
        print("Cinema:")
        print("  ",end=" ")
        
        for a in range(1,self.c+1):
            print(a,end=" ")
            
        for i in range(self.r):
            print("\n",i+1,end=" ")
            
            for j in range(self.c):
                print (self.matrix[i][j],end=" ")
        print("\n")
        
        self.menu()
        
    def buy(self):
        
        
        r=int(input("Enter row "))
        c=int(input("Enter coloumn "))
        
        if 0<r<=self.r and 0<c<=self.c:
            
            if self.matrix[r-1][c-1]=="B":
                
                print ("Sorry..Seat already booked!!")
                c=input("Do you want to book another seat (y\n)")
                
                if c=="y" or c=="Y":
                    self.buy()
                else:
                    return
            else:
                if  self.total_seats<60:
                    price=10
            
                elif r <= self.r//2:
                    price=10
        
                else:
                  price=8
                choice=input("The ticket price is {}$ \nDo you want to continue(y/n)".format(price))
                
                if choice=="y" or choice=="Y":
            
                    self.name=input("Enter your name: ")
                    self.age=input("Enter your age: ")
                    self.gen=input("Enter your Gender: ")
                    self.mob=input("Enter your Mob No: ")
                    self.matrix[r-1][c-1]="B"
                    self.total_income+=price
                    print("Your seat has been booked..")
                    self.userdict[(r,c)]=[self.name,self.age,self.gen,self.mob,price]
                    c=input("Do you want to book another seat (y/n)")
                    
                    if c=="y" or c=="Y":
                        self.buy()
                    else:
                        self.menu()
                else:
                    print("booking cancelled")
                    self.menu()
        else:
            print("Enter a valid seat number!!")
            self.buy()
        self.menu()
    
    def booking_details(self):
        
        r=int(input("enter your row number" ))
        c=int(input("enter your coloumn number "))
        
        if 0<r<=self.r and 0<c<=self.c:
            
            if self.matrix[r-1][c-1]=="B":
                details=["Name","Age","Gender","Mob no","Ticket Price"]
                for i in range(5):
                     print(details[i]," : ",self.userdict[(r,c)][i])
            else:
                print("Seat not booked")
        else:
            print("Invalid Seat no!!\n Enter Again")
            self.booking_details()
        self.menu()    
       
    def statistics(self):
        
        print("Enter your choice \n1) Number of purchased tickets\n2) Percentage of tickets booked\n3) Total income \n4)Main menu")
        choice=input("\nEnter Your Choice:")
        if choice=="1":
            print(len(self.userdict))
            self.statistics()
        elif choice=="2":
            print((len(self.userdict)*100)/self.total_seats)
            self.statistics()
        elif choice=="3":
            print(self.total_income)
            self.statistics()
        elif choice=="4":
            self.menu()
        else:
            print("Wrong Input!!")
            self.statistics()
    
```
